```python
import pandas as pd
from pathlib import Path

tsv = Path("/mnt/c/EEGFMRI_PIPELINE/templateflow/tpl-MNI152NLin2009cAsym/"
           "tpl-MNI152NLin2009cAsym_atlas-Schaefer2018_desc-200Parcels7Networks_dseg.tsv")

nii = Path("/mnt/c/EEGFMRI_PIPELINE/templateflow/tpl-MNI152NLin2009cAsym/"
           "tpl-MNI152NLin2009cAsym_res-01_atlas-Schaefer2018_desc-200Parcels7Networks_dseg.nii.gz")

out_txt = nii.with_suffix("").with_suffix(".txt")  # drop .nii.gz -> .txt

df = pd.read_csv(tsv, sep="\t")

# TemplateFlow dseg TSVs typically have "index" and "name" (but we handle variations)
idx_col = None
name_col = None
for c in df.columns:
    cl = c.lower()
    if idx_col is None and cl in ("index", "id", "label", "value"):
        idx_col = c
    if name_col is None and "name" in cl:
        name_col = c

if idx_col is None or name_col is None:
    # fallback: first two columns
    idx_col = df.columns[0]
    name_col = df.columns[1]

with open(out_txt, "w", encoding="utf-8") as f:
    for _, r in df.iterrows():
        f.write(f"{int(r[idx_col])} {str(r[name_col])}\n")

print("Wrote:", out_txt)
```
