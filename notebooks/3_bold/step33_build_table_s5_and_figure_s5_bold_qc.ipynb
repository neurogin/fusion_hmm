```python
# Step 0. Import Python modules and locate the Stage-3 helper module

import sys
from pathlib import Path

STAGE3_DIR = Path.cwd()
candidate = Path.cwd() / "notebooks" / "3_bold"
if not (STAGE3_DIR / "stage3_bold_summary_helpers.py").exists() and candidate.exists():
    STAGE3_DIR = candidate

if not (STAGE3_DIR / "stage3_bold_summary_helpers.py").exists():
    raise FileNotFoundError(
        f"Stage-3 helper not found: {STAGE3_DIR / 'stage3_bold_summary_helpers.py'}"
    )

if str(STAGE3_DIR.resolve()) not in sys.path:
    sys.path.insert(0, str(STAGE3_DIR.resolve()))

from stage3_bold_summary_helpers import build_table_s5_and_figure_s5_bold_qc
```

```python
# Step 3. Build Table S5 and the explicit Figure S5 reconstruction

table_s5, figure_info = build_table_s5_and_figure_s5_bold_qc(
    out_root=EXPORT_ROOT,
    table_s5_csv=TABLE_S5_CSV,
    figure_s5_png=FIGURE_S5_PNG,
    manual_notes_by_run=MANUAL_NOTES_BY_RUN,
)

table_s5
```
