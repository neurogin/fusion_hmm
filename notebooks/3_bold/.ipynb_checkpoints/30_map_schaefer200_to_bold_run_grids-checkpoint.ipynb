# 30 Map Schaefer200 To BOLD Run Grids

## What This Notebook Does

This public-facing Stage-3 notebook maps the Schaefer2018 200-parcel atlas onto each run's BOLD voxel grid and writes atlas-preservation QC outputs.

It preserves the standalone atlas-on-BOLD-grid branch from the original Stage-3 workflow. That branch is useful for atlas-preservation checks and overlay provenance, but it is **not** treated here as the authoritative atlas producer for the parcel-export path.

## Inputs

- fMRIPrep BOLD reference images
- matching brain masks when available
- a canonical Schaefer2018 atlas NIfTI and TSV
- optional raw BIDS JSON sidecars for metadata logging

## Outputs

- per-run atlas-on-BOLD-grid NIfTI files
- optional atlas overlay PNGs
- `qc_atlas_on_boldgrid_summary.csv`


```python
import sys
from pathlib import Path

HELPER_DIR = Path.cwd()
if not (HELPER_DIR / "stage3_bold_export_helpers.py").exists():
    candidate = Path.cwd() / "notebooks" / "3_bold"
    if candidate.exists():
        HELPER_DIR = candidate
if str(HELPER_DIR.resolve()) not in sys.path:
    sys.path.insert(0, str(HELPER_DIR.resolve()))

from stage3_bold_export_helpers import run_atlas_on_boldgrid

# User-editable inputs
FMRI_ROOT = Path(r"/mnt/c/EEGFMRI/hmm/R01_rerun/01_raw/fmri_bids")
OUT_ROOT = Path(r"/mnt/c/EEGFMRI/hmm/R01_rerun/02_derivatives/bold_parcel/atlas_on_boldgrid")
ATLAS_NII = Path(r"/mnt/c/EEGFMRI_PIPELINE/templateflow/tpl-MNI152NLin2009cAsym/tpl-MNI152NLin2009cAsym_res-01_atlas-Schaefer2018_desc-200Parcels7Networks_dseg.nii.gz")
ATLAS_TSV = Path(str(ATLAS_NII).replace("_dseg.nii.gz", "_dseg.tsv"))
SPACE_TAG = "space-MNI152NLin2009cAsym"
OVERWRITE = True
RAW_BIDS_ROOT = None
MAKE_OVERLAY_PNGS = True
```

```python
qc_df = run_atlas_on_boldgrid(
    fmri_root=FMRI_ROOT,
    out_root=OUT_ROOT,
    atlas_nii=ATLAS_NII,
    atlas_tsv=ATLAS_TSV,
    space_tag=SPACE_TAG,
    overwrite=OVERWRITE,
    raw_bids_root=RAW_BIDS_ROOT,
    make_overlay_pngs=MAKE_OVERLAY_PNGS,
)

qc_df.head()
```
