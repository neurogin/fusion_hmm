# Step 51 Run LOSO Shortlist Stability Checks

## What This Notebook Does

This is the public Stage-5 shortlist notebook for fusion-HMM model-order selection.

It preserves the shortlist-stability workflow that:
- re-fits shortlisted K values across LOSO folds,
- performs cross-fold state matching,
- writes transition-stability and signature-similarity summaries,
- and writes the matched fold-summary tables used later for Figure 2 and Table S8.
- This stage consumes the canonical Stage-4 alignment branch: `intermediate`, `nolags`, `minlen15`, where `intermediate` denotes the manuscript’s middle-ground TR-retention policy between earlier lenient and strict variants.

## When To Run It

Run this notebook after `step50_run_loso_k_sweep_model_selection.ipynb`.

The public manuscript-facing shortlist comparison is:
- `K_LIST = [3, 5]`

Higher-K candidates remain part of the earlier screening-stage provenance, but they are not the main public comparison carried forward here.

## Manuscript Linkage

- Main Methods 2.5.1
- Supplementary Methods 1.7
- Figure 2 support
- Table S8 support

## Inputs Expected

- the same Stage-4 canonical no-lag `minlen15` segment manifest used by the K-sweep stage
- the segment `.npy` files referenced by that manifest

## Outputs Written

- per-K subfolders such as `K03/` and `K05/`
- `state_matching_scores.tsv`
- `sim_matrix_signature.tsv`
- `sim_matrix_A.tsv`
- `A_mean.tsv`
- `A_std.tsv`
- `fold_summaries_table_matched.tsv`
- stability plots and per-fold logs

## Environment Notes

This stage depends on:
- `tensorflow`
- `osl_dynamics`
- GPU-sensitive execution for the main workflow

The preserved source notebook still uses XLA-off settings, memory-growth or GPU-cap handling, and chunk/resume behavior. CPU-only execution is possible in code, but it is likely slower and less practical for the full shortlist comparison.

## Interpretation Note

The earlier K-sweep screening stage still left higher-K candidates visible, including `K=12` and other local minima. The main manuscript-facing shortlist comparison is centered on `K=3` and `K=5`, because the higher-K solutions mainly introduced unwanted extra states rather than a cleaner shared architecture. The final manuscript choice is `K=3`.


```python
from stage5_hmm_selection_helpers import run_wrapped_shortlist_source
```

```python
# Step 1. User-editable roots and preserved shortlist settings
#
# Set SEGMENTS_ROOT to the Stage-4 final output root that contains the
# `intermediate/` subfolder built in Stage 4.

SEGMENTS_ROOT = Path("<SET_SEGMENTS_ROOT>")
SHORTLIST_OUTPUT_ROOT = Path("<SET_SHORTLIST_OUTPUT_ROOT>")

DATA_VARIANT = "intermediate"
FEATURE_MODE = "nolags"
MINLEN = 15

# Public manuscript-facing shortlist comparison.
K_LIST = [3, 5]

# Optional explicit manifest override.
MANIFEST_TSV = None

# Chunk/resume and debug controls for long runs.
MAX_NEW_FOLDS_PER_RUN = 1
FORCE_RERUN_HELDOUTS = []
DEBUG_SUBJECTS = None
DEBUG_SEEDS = None

# Keep None to use TensorFlow memory-growth mode.
GPU_MEMORY_LIMIT_MB = None

FINAL_ROOT = SEGMENTS_ROOT / DATA_VARIANT
SHORTLIST_RUN_DIR = SHORTLIST_OUTPUT_ROOT / f"PipelineD_C2_{DATA_VARIANT}_{FEATURE_MODE}_minlen{MINLEN}"
SOURCE_NOTEBOOK = STAGE5_DIR / "R01_PipelineD_C2_LOSO_stability_shortlist_REWRITE_v3_twostage_configfix.ipynb"
```

```python
# Step 2. Validate the configured inputs and print a short run summary

def assert_configured_path(path_value, label, must_exist=False):
    path_str = str(path_value)
    if "<SET_" in path_str:
        raise ValueError(f"{label} is still using a placeholder path. Edit this notebook before running.")
    if must_exist and not Path(path_value).exists():
        raise FileNotFoundError(f"{label} does not exist: {path_value}")

assert_configured_path(SEGMENTS_ROOT, "SEGMENTS_ROOT", must_exist=True)
assert_configured_path(SHORTLIST_OUTPUT_ROOT, "SHORTLIST_OUTPUT_ROOT")

if not SOURCE_NOTEBOOK.exists():
    raise FileNotFoundError(f"Preserved source notebook not found: {SOURCE_NOTEBOOK}")

print("Stage 5 / Step 51: shortlist stability and state matching")
print(f"  Stage-4 segment root:   {FINAL_ROOT}")
print(f"  Output root:            {SHORTLIST_RUN_DIR}")
print(f"  Feature mode:           {FEATURE_MODE}")
print(f"  Minimum segment len:    {MINLEN} TR")
print(f"  Public shortlist:       {K_LIST}")
print(f"  GPU memory cap (MB):    {GPU_MEMORY_LIMIT_MB}")
print("  Packages needed:        tensorflow, osl_dynamics, numpy, pandas, matplotlib, scipy")
```

```python
# Step 3. Run the preserved shortlist notebook through the public wrapper

run_wrapped_shortlist_source(
    source_notebook=SOURCE_NOTEBOOK,
    final_root=FINAL_ROOT,
    out_root=SHORTLIST_RUN_DIR,
    k_list=K_LIST,
    data_variant=DATA_VARIANT,
    feature_mode=FEATURE_MODE,
    minlen=MINLEN,
    manifest_tsv=MANIFEST_TSV,
    max_new_folds_per_run=MAX_NEW_FOLDS_PER_RUN,
    gpu_memory_limit_mb=GPU_MEMORY_LIMIT_MB,
    force_rerun_heldouts=FORCE_RERUN_HELDOUTS,
    debug_subjects=DEBUG_SUBJECTS,
    debug_seeds=DEBUG_SEEDS,
)
```
