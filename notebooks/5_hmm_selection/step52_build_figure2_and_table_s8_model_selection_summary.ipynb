# Step 52 Build Figure 2 and Table S8 Model-Selection Summary

## What This Notebook Does

This is the public Stage-5 summary notebook for the model-selection stage.

It rebuilds the manuscript-facing summary assets by:
- reading the broad K-sweep outputs from Step 50,
- reading the per-K shortlist outputs from Step 51,
- building a compact model-selection table,
- and rebuilding the composite summary figure used for Figure 2 support.

## When To Run It

Run this notebook after:
- `step50_run_loso_k_sweep_model_selection.ipynb`
- `step51_run_loso_shortlist_stability_checks.ipynb`

## Manuscript Linkage

- Figure 2
- Table S8

## Inputs Expected

- `summary_byK_selected.tsv`
- `K_selection_recommendation.json`
- per-K shortlist outputs such as `K03/state_matching_scores.tsv`
- per-K shortlist outputs such as `K03/fold_summaries_table_matched.tsv`

## Outputs Written

- `fusion_hmm_model_selection_summary.png`
- `fusion_hmm_K_selection_compact_table.csv`
- `model_selection_summary_inputs.json`

## Environment Notes

This notebook does not rerun the main HMM fitting. It only depends on the saved Stage-5 outputs plus the standard Python summary stack used here (`numpy`, `pandas`, and `matplotlib`).

## Interpretation Note

This notebook keeps the two-stage model-selection story explicit:
- first screen a broad `K = 2..12` sweep,
- then compare the manuscript-facing shortlist `K=3` versus `K=5`,
- and carry the final manuscript choice as `K=3`.

It does not present the final `K=3` result as if it came from one fully automatic rule alone.


```python
# Step 0. Import Python modules and locate the active Stage-5 helper module

import sys
from pathlib import Path

STAGE5_DIR = Path.cwd()
candidate = Path.cwd() / "notebooks" / "5_hmm_selection"
if not (STAGE5_DIR / "stage5_hmm_selection_helpers.py").exists() and candidate.exists():
    STAGE5_DIR = candidate

if not (STAGE5_DIR / "stage5_hmm_selection_helpers.py").exists():
    raise FileNotFoundError(
        f"Stage-5 helper not found: {STAGE5_DIR / 'stage5_hmm_selection_helpers.py'}"
    )

if str(STAGE5_DIR.resolve()) not in sys.path:
    sys.path.insert(0, str(STAGE5_DIR.resolve()))

from stage5_hmm_selection_helpers import build_figure2_and_table_s8_summary
```

```python
# Step 1. User-editable input and output roots

MODEL_SELECTION_ROOT = Path("<SET_MODEL_SELECTION_ROOT>")
SHORTLIST_OUTPUT_ROOT = Path("<SET_SHORTLIST_OUTPUT_ROOT>")
SUMMARY_OUTPUT_ROOT = Path("<SET_SUMMARY_OUTPUT_ROOT>")

# Main manuscript-facing comparison used in the compact summary.
COMPARE_KS = (3, 5)

# Keep screening-stage K values visible in the figure annotations.
ANNOTATE_KS = (3, 5, 9, 12)

# Keep a higher-K screening-stage reference in the compact decision table.
DECISION_TABLE_KS = (3, 5, 12)

FIGURE_NAME = "fusion_hmm_model_selection_summary.png"
TABLE_NAME = "fusion_hmm_K_selection_compact_table.csv"
FIGSIZE = (12, 10)
DPI = 200
```

```python
# Step 2. Validate the configured inputs and print a short run summary

def assert_configured_path(path_value, label, must_exist=False):
    path_str = str(path_value)
    if "<SET_" in path_str:
        raise ValueError(f"{label} is still using a placeholder path. Edit this notebook before running.")
    if must_exist and not Path(path_value).exists():
        raise FileNotFoundError(f"{label} does not exist: {path_value}")

assert_configured_path(MODEL_SELECTION_ROOT, "MODEL_SELECTION_ROOT", must_exist=True)
assert_configured_path(SHORTLIST_OUTPUT_ROOT, "SHORTLIST_OUTPUT_ROOT", must_exist=True)
assert_configured_path(SUMMARY_OUTPUT_ROOT, "SUMMARY_OUTPUT_ROOT")

print("Stage 5 / Step 52: build Figure 2 and Table S8 summary outputs")
print(f"  K-sweep input root:     {MODEL_SELECTION_ROOT}")
print(f"  Shortlist input root:   {SHORTLIST_OUTPUT_ROOT}")
print(f"  Summary output root:    {SUMMARY_OUTPUT_ROOT}")
print(f"  Main comparison:        {COMPARE_KS}")
print(f"  Figure annotations:     {ANNOTATE_KS}")
print(f"  Decision-table Ks:      {DECISION_TABLE_KS}")
```

```python
# Step 3. Build the Stage-5 summary figure and compact decision table

outputs = build_figure2_and_table_s8_summary(
    ksweep_output_dir=MODEL_SELECTION_ROOT,
    stability_output_dir=SHORTLIST_OUTPUT_ROOT,
    summary_output_dir=SUMMARY_OUTPUT_ROOT,
    compare_ks=COMPARE_KS,
    decision_table_ks=DECISION_TABLE_KS,
    annotate_ks=ANNOTATE_KS,
    figure_name=FIGURE_NAME,
    table_name=TABLE_NAME,
    figsize=FIGSIZE,
    dpi=DPI,
)

outputs
```
