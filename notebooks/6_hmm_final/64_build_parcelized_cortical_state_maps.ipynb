# 64 Build Parcelized Cortical State Maps

## What This Notebook Does

This is the public Stage-6 surface-map notebook for the saved final `K = 3` fusion HMM.

It preserves the original parcelized cortical map workflow that:
- loads the saved state-signature BOLD correlation summaries
- projects the Schaefer-200 parcel values to fsaverage surfaces
- builds a reference atlas map
- and exports parcelized cortical maps for the dominant/reference state and the two contrast states
