# Step 62 Reconstruct BOLD State Networks and Ranked Contrasts

## What This Notebook Does

This is the public Stage-6 BOLD-state reconstruction notebook for the saved final `K = 3` fusion HMM.

It preserves the original physiology-style workflow that:
- reconstructs state-wise parcel-by-parcel BOLD correlation matrices from the saved final-model signature file
- attaches Schaefer parcel labels and seven-network labels
- summarizes network block structure
- ranks the strongest BOLD contrasts relative to the chosen reference state
- and writes the summary tables used downstream for manuscript interpretation

## When To Run It

Run this notebook after `step60_fit_final_k3_fusion_hmm.ipynb` has completed successfully.

- the saved Stage-6 final-model output folder
- Schaefer atlas label files from TemplateFlow or the matching Brainstorm text export
- optional `preproc_params.npz` if you also want the preserved covariance backprojection check

## Outputs Written

- state-wise parcel and network summary figures
- ranked network and parcel contrast tables
- `state_summary_table.tsv`

## Environment Notes

This notebook works from saved final-fit outputs rather than rerunning the fit. It still expects the Python scientific stack used by the preserved notebook, plus access to the Schaefer label files under TemplateFlow or an equivalent exported label file.

## Scientific Notes

- The preserved notebook treats the reference state explicitly.

- If the saved final FO vector makes `S2` dominant, the notebook will report `S2` as the dominant and reference state rather than assuming it in advance.
