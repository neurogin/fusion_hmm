# Step 60 Fit Final K3 Fusion HMM

## What This Notebook Does

This is the public Stage-6 final-fit notebook for the manuscript's chosen full-data fusion HMM.

It preserves the original final-model workflow that:
- loads the canonical retained Stage-4 segment dataset,
- applies run-wise normalization and global PCA,
- screens multiple random seeds,
- refits the top seeds,
- chooses the final non-collapsed `K = 3` solution,
- and writes the saved model artifacts used by later review and reconstruction notebooks.

## When To Run It

Run this notebook after the canonical Stage-4 retained dataset has been built.

Run this notebook before:
- `step61_review_final_k3_fit_qc_and_state_dynamics.ipynb`
- `step62_reconstruct_bold_state_networks_and_ranked_contrasts.ipynb`
- `step63_reconstruct_crossmodal_state_blocks_and_ranked_contrasts.ipynb`
- `step64_build_parcelized_cortical_state_maps.ipynb`
- `step65_optional_export_figure4_figure5_panels.ipynb`

## Environment Notes

This stage depends on:
- `tensorflow`
- `osl_dynamics`
- GPU-sensitive execution for the main workflow

The preserved source notebook keeps XLA-off settings, GPU memory-growth or GPU-cap handling, and chunk-safe fitting behavior. CPU-only execution exists in code, but it is likely slower and less practical for the full final fit.

## Scientific Notes

- This stage depends directly on the Stage-4 retained dataset.
- Stage 5 contributes the scientific decision to carry `K = 3` forward, but it is not a file dependency here.
- The public default stays aligned to the frozen manuscript dataset:
  - `DATA_VARIANT = "intermediate"`
  - `FEATURE_MODE = "nolags"`
  - `MINLEN = 15`
  - `K_FINAL = 3`


```python
# Step 0. Import Python modules and locate the Stage-6 helper folder
 
 import sys
 from pathlib import Path
 
 STAGE6_DIR = Path.cwd()
 if not (STAGE6_DIR / "helpers" / "stage6_hmm_final_helpers.py").exists():
     candidate = Path.cwd() / "notebooks" / "6_hmm_final"
     if candidate.exists():
         STAGE6_DIR = candidate
 
 HELPER_DIR = STAGE6_DIR / "helpers"
 if not (HELPER_DIR / "stage6_hmm_final_helpers.py").exists():
     raise FileNotFoundError(f"Stage-6 helper not found: {HELPER_DIR / 'stage6_hmm_final_helpers.py'}")
 
 if str(HELPER_DIR.resolve()) not in sys.path:
     sys.path.insert(0, str(HELPER_DIR.resolve()))
 
 from stage6_hmm_final_helpers import ensure_configured_path, run_stage6_final_fit_source
```

```python
# Step 1. User-editable roots and preserved final-fit settings
#
# Set SEGMENTS_ROOT to the Stage-4 final output root that contains the
# canonical `intermediate/` branch built in Stage 4.

SEGMENTS_ROOT = Path("<SET_SEGMENTS_ROOT>")
FINAL_MODEL_ROOT = Path("<SET_FINAL_MODEL_ROOT>")

DATA_VARIANT = "intermediate"
FEATURE_MODE = "nolags"
MINLEN = 15
K_FINAL = 3

# Optional explicit manifest override.
# Leave this as None to use the standard Stage-4 manifest location.
MANIFEST_TSV = None

# Set this to None to rely on TensorFlow memory-growth instead of a hard cap.
GPU_MEMORY_LIMIT_MB = None

# Keep the preserved decoded outputs enabled unless you have a strong reason to change them.
SAVE_GAMMA = True
DO_VITERBI = True

SEGMENT_BRANCH_ROOT = SEGMENTS_ROOT / DATA_VARIANT
FINAL_MODEL_OUTPUT_ROOT = FINAL_MODEL_ROOT / f"PipelineE_final_K{K_FINAL:02d}_{DATA_VARIANT}_{FEATURE_MODE}_minlen{MINLEN}"
SOURCE_NOTEBOOK = STAGE6_DIR / "R01_PipelineE_full_model.ipynb"
```

```python
# Step 2. Validate the configured inputs and print a short run summary

ensure_configured_path(SEGMENTS_ROOT, "SEGMENTS_ROOT", must_exist=True)
ensure_configured_path(FINAL_MODEL_ROOT, "FINAL_MODEL_ROOT")

if not SOURCE_NOTEBOOK.exists():
    raise FileNotFoundError(f"Preserved source notebook not found: {SOURCE_NOTEBOOK}")

print("Stage 6 / Step 60: final full-data K=3 fusion HMM fit")
print(f"  Stage-4 segment root:     {SEGMENT_BRANCH_ROOT}")
print(f"  Output directory:         {FINAL_MODEL_OUTPUT_ROOT}")
print(f"  Feature mode:             {FEATURE_MODE}")
print(f"  Minimum segment length:   {MINLEN} TR")
print(f"  Final chosen K:           {K_FINAL}")
print(f"  GPU memory cap (MB):      {GPU_MEMORY_LIMIT_MB}")
print(f"  Save gamma outputs:       {SAVE_GAMMA}")
print(f"  Save Viterbi outputs:     {DO_VITERBI}")
print("  Packages needed:          tensorflow, osl_dynamics, numpy, pandas")
```

```python
# Step 3. Run the preserved final-fit notebook through the public wrapper

run_stage6_final_fit_source(
    source_notebook=SOURCE_NOTEBOOK,
    final_root=SEGMENT_BRANCH_ROOT,
    out_root=FINAL_MODEL_OUTPUT_ROOT,
    manifest_tsv=MANIFEST_TSV,
    k_final=K_FINAL,
    data_variant=DATA_VARIANT,
    feature_mode=FEATURE_MODE,
    minlen=MINLEN,
    gpu_memory_limit_mb=GPU_MEMORY_LIMIT_MB,
    save_gamma=SAVE_GAMMA,
    do_viterbi=DO_VITERBI,
)
```
